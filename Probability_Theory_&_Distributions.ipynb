# Binomial Distribution (Discrete Distribution)
* A probability distribution that summarizes the likelihood that a value will take one of two independent values under a given set of parameters or assumptions
* has parameters **n** and **p** where **n** is the number of runs and **p** is the probability of success
* samples is an array with the size of 1,000. Each number in the array represents how many times, out of 10 tosses, we get heads
* Example: Coin Toss

```
# n - number of tosses
# p - probaility of heads
# size - number of repeats

n = 10
#assuming we have a fair coin
p = 0.5
size = 1000

samples = np.random.binomial(n,p,size)
```

```
#use Counter to find the occurence of each number in samples
head_occu_count = Counter(samples)
head_occu_count
```

```
#calculate probability of head as occurance/size
prob_head = [val/size for val in head_occu_count.values()]
prob_head
```

```
#use a scatter plot to show the probability distribution
plt.scatter(x=head_occu_count.keys(), y=prob_head)
plt.xlabel('number of heads out of 10 tosses')
plt.ylabel('probability')

# show the plot
plt.show()

#use can use the plot to find the probability of x heads 
```

# Normal Distribution (Continuous Distribution)
* Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. 
* In graph form, normal distribution will appear as a bell curve
* Normal distributions are symmetrical, but not all symmetrical distributions are normal
* use numpy to generate a Normal Distribution using normal() 

* For a continuous random variable, the probability density function provides the height, or value, of the function for any particular value of x. It does not directly give us the probability of the random variable taking on a specific value. 
* However, the area under the graph corresponding to a certain interval provides the probability that the variable will take on a value within that interval.

```
#generate a Normal Distrinution using numpy
# loc = mean (“a centre”) of the distribution.
# scale = standard deviation (a spread or “a width”) of the distribution.
# the normal distribution below has mean 0 and standard distribution of 1 (Standard Normal Distribution)

samples = np.random.normal(loc=0, scale=1, size = 1000000)
```

```
#plot the Probability Density Function (PDF) using a histogram
plt.hist(samples, density=True, histtype="step", bins=100)

# show the plot
plt.rcParams["figure.figsize"] = (10, 8)
plt.show()
```

```
#use different parameters for Standard Deviation to explore how the PDF changes
samples_std1 = np.random.normal(20, 1, size=100000)
samples_std4 = np.random.normal(20, 4, size=100000)
samples_std12 = np.random.normal(20, 12, size=100000)

# make histograms
plt.hist(samples_std1, density = True, histtype="step", bins=100)
plt.hist(samples_std4, density = True, histtype="step", bins=100)
plt.hist(samples_std12, density = True, histtype="step", bins=100)

# make a legend, set limits, and show the plot
plt.rcParams["figure.figsize"] = (12, 10)
plt.legend(('std = 1', 'std = 3', 'std = 10'))
plt.ylim(-0.01, 0.42)
plt.show() 

# From the plot below, the higher the std value is, the more spread out the distribution
```

```
#use different parameters for Mean to explore how the PDF changes
samples_mn10 = np.random.normal(10, 3, size=100000)
samples_mn20 = np.random.normal(20, 3, size=100000)
samples_mn30 = np.random.normal(30, 3, size=100000)

# make histograms
plt.hist(samples_mn10, density = True, histtype="step", bins=100)
plt.hist(samples_mn20, density = True, histtype="step", bins=100)
plt.hist(samples_mn30, density = True, histtype="step", bins=100)

# make a legend, set limits, and show the plot
plt.rcParams['figure.figsize'] = (12,10)
plt.legend(('mean = 10', 'mean = 20', 'mean = 30'))
plt.ylim(-0.01, 0.150)
plt.show()

#As shown by the plot below, the width of our distributions did not change but the distributions shifted along the x-axis
```
